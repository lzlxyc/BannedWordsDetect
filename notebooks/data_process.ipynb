```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 设置字体为 SimHei（黑体）以支持中文
plt.rcParams['font.sans-serif'] = ['SimHei']
# 避免负号显示为方块
plt.rcParams['axes.unicode_minus'] = False



df = pd.read_csv('../DATA/train_all.csv')

df['text_len'] = df['文本'].str.len()
df = df.sort_values('text_len', ascending=False)
```

```python
# 箱线图展示分布
plt.figure(figsize=(10, 6))
sns.boxplot(x='类别', y='text_len', data=df)
plt.title('文本长度按类别分布')
plt.xticks(rotation=45)
plt.show()

# 小提琴图展示分布密度
plt.figure(figsize=(10, 6))
sns.violinplot(x='类别', y='text_len', data=df)
plt.title('文本长度按类别分布密度')
plt.xticks(rotation=45)
plt.show()
```

```python
import matplotlib.pyplot as plt

# 绘制直方图
df['文本'].str.len().hist(bins=50)
plt.xlabel('文本长度')
plt.ylabel('频数')
plt.title('文本长度分布')
plt.show()
```
